```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf

from statsmodels.tsa.seasonal import seasonal_decompose, STL
```

CRIPTO - (ETC-USD)

```python
dfETC = pd.read_csv('/content/ETC-USD.csv')
dfETC.head()
```

```python
# Converte a coluna Date para o tipo de dados Datetime do Pandas
dfETC['Date'] = pd.to_datetime(dfETC['Date'])

# Calcular os retornos
dfETC['Retorno'] = dfETC['AdjClose'].pct_change()

# Criar uma figura com dois (gráficos)
fig, ax = plt.subplots(2, 1, figsize=(12, 8))

# Gráfico de preços
ax[0].plot(dfETC['Date'], dfETC['AdjClose'], label='Preços')
ax[0].set_title('Preços ETC')
ax[0].legend()

# Gráfico de retornos
ax[1].plot(dfETC['Date'], dfETC['Retorno'], label='Retornos', color='orange')
ax[1].set_title('Retornos ETC')
ax[1].legend()


# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

```python
# Criar uma variavel precos e atribuir valor da coluna AdjClose
# Criar uma variavel retornos e atribuir valor da coluna retorno
precos = dfETC['AdjClose']
retornos = dfETC['Retorno']

# Realizar o teste ADF nos preços e imprimi os valores obtidos
resultado_precos = adfuller(precos.dropna())
print(f"ADF Statistic para preços: {resultado_precos[0]}")
print(f"p-value para preços: {resultado_precos[1]}")

# Realizar o teste ADF nos retornos e imprimi os valores obtidos
resultado_retornos = adfuller(retornos.dropna())
print(f"ADF Statistic para retornos: {resultado_retornos[0]}")
print(f"p-value para retornos: {resultado_retornos[1]}")

#ADF Statistic: Esta é a estatística do teste. Quanto mais negativo for este valor, mais forte é a rejeição da hipótese nula de que existe uma raiz unitária na série temporal.
#p-value: Este é o valor-p do teste. Se o valor-p for menor que 0.05, a série temporal é estacionária.
```

O valor de p para os preços é muito pequeno e praticamente igual a zero, ou seja, menor que 0,05 e neste caso podemos rejeitar a hipotese nula. O valor de p para o retorno é maior que 0,05 e neste caso não podemos rejeitar.

```python
# Correlograma para preços
plot_acf(precos.dropna())
plt.title('Correlograma para Preços ETC')
plt.show()

# Correlograma para retornos
plot_acf(retornos.dropna())
plt.title('Correlograma para Retornos ETC')
plt.show()
```
